## 25ページ

プログラム中でキを求めている。ラグビーでみる横に流れていくパスのように感じる。

```python
Akibi = [5, 3, 4]
buinsu = 3
tantou = 0
for buin in range(1, buinsu):
    if Akibi[buin] < Akibi[tantou]:
        tantou = buin
print("次の工芸品の担当は部員", tantou, "です。")
```

続けて, 部員数を変更した場合を考察している。

```python
Akibi = [5, 6, 4, 4, 4]
buinsu = 5
tantou = 0
buin = 0
for buin in range(1, buinsu):
    if Akibi[buin] < Akibi[tantou]:
        tantou = buin
print("次の工芸品の担当は部員", tantou, "です。")
```

## 26ページ

```python
Nissu = [4, 1, 3, 1, 3, 4, 2, 4, 3]
kougeihinsu = 9
Akibi = [0, 0, 0]
buinsu = 3
for kougeihin in range(kougeihinsu):
    tantou = 0
    for buin in range(1, buinsu):
       if Akibi[buin] < Akibi[tantou]:
          tantou = buin
    print("工芸品", kougeihin, " ... ",
          "部員", tantou, ":",
          Akibi[tantou], "日目 ～",
          Akibi[tantou]+Nissu[kougeihin]-1, "日目")
    Akibi[tantou]=Akibi[tantou]+Nissu[kougeihin]
```

# **まとめ**

Pythonで起こしたプログラムとよく見比べてほしい。ほぼ一緒であることに気づくであろう。今回の問題は繰返し処理が2重になって, その一番内側に条件分岐がある。標準的な問題である。配列（リスト）を使用しているが、その値を上書きすることはしていない（追試問題では上書きあり）。標準から簡単なレベルと考える。だからこそ, 添字を1からとし, 受験生を惑わせたのではないか。
