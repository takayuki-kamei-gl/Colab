図1.1 目標の金額ちょうどになる最小の硬貨枚数を計算する関数「枚数()」「Maisu(kingaku)」

````python
def Maisu(kingaku):
    Kouka = [1, 5, 10, 50, 100]
    maisu = 0
    nokori = kingaku
    for i in range(4, -1, -1):
        maisu = maisu + nokori//Kouka[i]
        nokori = nokori % Kouka[i]
    return(maisu)
````

図2 最小交換硬貨枚数を求めるプログラム

````python
kakaku = 46
min_maisu = 100
for tsuri in range(100):
    shiharai = kakaku + tsuri
    maisu  = Maisu(shiharai)+Maisu(tsuri)
    if maisu < min_maisu:
        min_maisu = maisu
print(min_maisu)
````
