問１ Pyhonは代入された値によって変数の「型」が決まります。

```python
#プログラム１
a = 1       #整数型（小数を含まない数値）
b = 5
print(a + b)
```

```python
#プログラム２
a = "1"     #文字列型
b = "5"
print(a + b)
```

```python
#プログラム７
a = 5
wa = 0
for i in range(1, 11):
    a = a + 3
    wa = wa + a
    print(a)
print(wa)
```

```python
#プログラム７(修正後)
a = 5
wa = a
for i in range(2, 11):
    a = a + 3
    wa = wa + a
print(wa)
```

以下は「」内を生成AIに入力し、作ってもらったプログラム（3行目以降）です。

```python
# 「初項が5、公差が3の数列の10項目までの和を求めるPythonプログラム」(by Copilot)

# 初項
first_term = 5
# 公差
common_difference = 3
# 項数
number_of_terms = 10

# 和を初期化
sum_of_series = 0

# for文を使って数列の各項を計算し、和に加える
for i in range(number_of_terms):
    term = first_term + i * common_difference
    sum_of_series += term

# 結果を出力
print(f"初項が{first_term}、公差が{common_difference}の数列の10項目までの和は{sum_of_series}です。")
```
