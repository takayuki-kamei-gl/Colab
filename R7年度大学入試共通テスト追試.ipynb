この部分を本解説では, ランダムに40人分の適切な数値を入れるように改良する。

````python
import random
ninzu = 40
for i in range(ninzu):
    Iremono[i] = random.randint(1,2)
    Shurui[i]  = random.randint(1,2)
    Keiryou[i] = random.randint(500,2000) # 500g ～ 2000g
````

````python
kanen = 0
funen = 0
for i in range(ninzu):
    if Iremono[i] == 1:
        gomi = Keiryou[i] - 350
    else:
        gomi = Keiryou[i]
    if Shurui[i] == 1:
        kanen = kanen + gomi
    else:
        funen = funen + gomi
print("可燃ごみの総重量は", kanen, "g")
print("不燃ごみの総重量は", funen, "g")
````
