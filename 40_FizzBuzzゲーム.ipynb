## 問１
図１と図２の違いを、言語化してみよう。

```python
# 図１のフローチャートをプログラム
for i in range(1,16):
    if(i % 3 == 0):
        print("Fizz")
    if(i % 5 == 0):
        print("Buzz")
    if(i % 15 == 0):
        print("FizzBuzz")
    else:
        print(i)
```

```python
# 図２のフローチャートをプログラム
for i in range(1,16):
    if(i % 15 == 0):
        print("FizzBuzz")
    elif(i % 3 == 0):
        print("Fizz")
    elif(i % 5 == 0):
        print("Buzz")
    else:
        print(i)
```

## 問2
問題文とプログラムの関係（問題の構成）を、作問者の気持ちになって考えてみよう。

```python
#プログラム1
for i in range(1,101):
    if i % 15 == 0:
        print("FizzBuzz")
    if i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
#プログラム2
for i in range(1, 101):
    if i % 15 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
#プログラム3
for i in range(1, 101):
    a = ""
    if i % 3 == 0:
        a = "Fizz"
    if i % 5 == 0:
        a = a + "Buzz"
    if a == "":
        a = i
    print(a)
```
